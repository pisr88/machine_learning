Manual calc

```python
import numpy as np
```

```python
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000,3250,3500,3750,4000,4250])
m = len(X1)
```

```python
X1
```

```python
X1 = X1.reshape(m,1)
```

```python
bias = np.ones((m,1))
```

```python
X = np.append(bias, X1, axis=1)
```

```python
X
```

Równanie normalne



```python
np.dot(X.T, X)
```

```python
L = np.linalg.inv(np.dot(X.T,X))
```

```python
P = np.dot(X.T, Y)
```

```python
np.dot(L, P)
```

SKLERN - auto

```python
from sklearn.linear_model import LinearxRegression

regression = LinearxRegression()
regression.fit(X1, Y)

print(regression.intercept_)
print(regression.coef_[0])
```
